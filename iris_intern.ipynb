```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ds= sns.load_dataset("iris")
```

```python
print("Number of Rows and colums : ")
print(ds.shape)
```

```python
print("Names of Columns:")
print(ds.columns)
```

```python
print("Showing first few rows: ")
print(ds.head())
```

```python
ds.info()
```

```python
ds.describe()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=ds, x='sepal_length', y='sepal_width', hue='species', style='species')
plt.title('Sepal Length vs Sepal Width')
plt.show()
```

```python
ds.hist(figsize=(10, 8), bins=15, edgecolor='black')
plt.suptitle('Distribution of All Features')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=ds)
plt.title('Identifying Outliers in Iris Features')
plt.show()
```
